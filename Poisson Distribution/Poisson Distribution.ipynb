### üñåÔ∏è Case Study: Paint Defect Monitoring in Tractor Manufacturing Using Poisson Distribution
üè≠ Company: AgroMotive Tractors Pvt. Ltd.

### üìå Problem Statement
AgroMotive's paint shop applies protective coatings on tractor bodies using robotic arms. Occasional defects‚Äîlike paint bubbles, runs, or dust inclusions‚Äîare noticed. The Quality Control (QC) team wants to:

Model the probability of defects in each hour of production.

Identify high-risk hours for intensive inspection.

Optimize manpower and reduce unnecessary inspections during low-risk hours.



### üìä Sample Data Summary
Duration: 40 continuous hours (approx. 5 shifts)

Column 1: Hour Number (1 to 40)

Column 2: Number of Paint Defects found

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
import pandas as pd

# Œª (mean defects per hour)
lambda_defects = 1.7

# Range of defects to evaluate & Check probabilities up to  6defects
k_values = np.arange(0, 6)
k_values
```

```python
# Poisson PMF
poisson_probs = poisson.pmf(k_values, mu=lambda_defects)

# Display Probabilities
print("Poisson Probabilities (Defects per Hour):")
for k, prob in zip(k_values, poisson_probs):
    print(f"P({k} defects) = {prob:.4f} ({prob*100:.1f}%)")
```

```python
# Plot
plt.bar(k_values, poisson_probs, color='orange', edgecolor='black')
plt.title("Poisson Distribution of Paint Defects per Hour")
plt.xlabel("Number of Defects")
plt.ylabel("Probability")
plt.grid(axis='y', linestyle='--')
plt.show()
```

 ### Interpretation
~75% of hours will have 0, 1, or 2 defects ‚Üí only spot checks needed.

~15% of hours have 3+ defects ‚Üí recommend full inspection.

High probability zone (1‚Äì2 defects) aligns well with historical trends.
